# Seinfield Analysis

#### Final Project - CUNY DATA 620 Web Analytics, Summer 2018
__Team:__ Andy Carson, Nathan Cooper, Walt Wells

## Overview:
Put on your urban sombreros and join us as we explore who among us is spongeworthy.  “Serenity Now!”  Seinfeld is a sitcom that ran on the US NBC network from 1989 to 1998, is universally beloved, and is generally considered to be one of the most important and influential sitcoms ever made.  In this project we will ‘double-dip’ with our data, reviewing full episode scripts and metadata associated with the cast, writers and directors of each episode to learn more about both the associated social networks and the text dialogue.   “That’s Gold, Jerry!   Gold!”  

## Proposal:
For our final project in DATA 620, we will study two aspects of the Seinfeld script corpus:  i) a social network analysis of characters, writers, and directors and ii) text analytics of the script dialogue.   First we will prepare a dataset based on scripts of each episode, where each script is a separate .html page.    The prepared dataset will organize the data by networks, while retaining maximum relevant information from the raw source to permit text analytics.   

We will then review social networks of writers, directors, cast, and then group characters according to their presence in scenes together.     In addition, we will analyze the script dialogue to try and find key phrases, and see if machine learning techniques can help predict the writers of a given episode.  

#### Key deliverables include:    
A prepared dataset, a dataset preparation notebook, a notebook on SNA,  a notebook with text analytics, a github repository with notebooks and any supplementary materials, a video presentation.

## Tasks:

### Data Preparation
__Data Source:__  Seinology - Citizen transcribed Seinfield scripts.
http://www.seinology.com/scripts.shtml

In order to facilitate SNA, we extracted additional information regarding scenes and dialogue order in scenes.   In addition, we've organized the data as separate tables as follows:

* T1:   Seinfield_Metadata.csv - KEY: Ep/SeasonNum, AirDate, EpTitle, Director
* T2:   Seinfield_Cast.csv - KEY Ep/SeasonNum,  Actor Name, Character
* T3:   Seinfield_Writers.csv - KEY Ep/SeasonNum,  Writer Name
* T4:   Seinfield_Dialogue.csv - KEY Ep/SeasonNum,  Character,  Text, Scene Location, Scene Dialogue Index
* T4:   Seinfield_DialogueSUBSET.csv - Same as Seinfeld Dialogue, but only keeps those entries with > 2 lines of dialogue across the entire corpus.   Cuts out most of the scrape noise from inconsistencies in transcription style.

> NOTE:   To help with both network dimensionality and parsing outliers, we will cut down the Dialogue data set to feature only those characters that had at least 2 lines of dialogue across the whole corpus.
